```python
ms_stock_prices = [25.4, 30, 40, 49, 59, 81.5, 100.6, 155.9, 220.9, 335.5]
apl_stock_prices = [20.4, 17.5, 25.7, 25.4, 27.4, 42.1, 52.5, 69.9, 127.5, 151.2]
years = np.arange(2012, 2022)
```

```python
plt.figure(figsize=(8,4))
plt.plot(years, ms_stock_prices)
plt.plot(years, apl_stock_prices)
```

```python
plt.figure(figsize=(8, 4))
plt.title('Comparison of Apple and Microsoft stocks')
plt.xlabel('years')
plt.ylabel('stock price (in Dollars)')
plt.plot(years, ms_stock_prices, label='Microsoft')
plt.plot(years, apl_stock_prices, label='Apple')
plt.xticks(years)
plt.legend()
plt.show()
```

**Ex: Write a function that returns the maximum value in a list**

```python
def find_maximum(list_of_numbers):
    maximum = list_of_numbers[0]
    for number in list_of_numbers:
        if number > maximum:
            maximum = number

    return maximum
```

```python
find_maximum([0,-2, 30, -45, 90])
```

```python
def find_maximum(list_of_numbers):
    sorted_list = sorted(list_of_numbers)
    return sorted_list[-1]
```

```python
find_maximum([0,-2, 30, -45, 90])
```
