```python
operator = '*'
a = 10
b = 8
if operator == '+':
    print(a + b)
else:
    if operator == '-':
        print(a - b)
    else:
        if operator == '*':
            print(a * b)
print('always print')
```

```python
operator = '/'
if operator == '+':
    print(a + b)
elif operator == '-':
    print(a - b)
elif operator == '*':
    print(a * b)
else:
    print(a / b)
```

```python
list_of_numbers = list(range(0, 100))
```

```python
#print the even numbers between 10 and 20 using a for loop
for even_number in list_of_numbers:
    if even_number >= 10 and even_number <= 20:
        print(even_number)
```

```python
for x in list_of_even_numbers:
    if x % 2 == 0 and x >= 10 and x <= 20:
        print(x)
```
