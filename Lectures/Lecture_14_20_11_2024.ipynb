```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ms_stock_prices = [25.4, 30, 40, 49, 59, 81.5, 100.6, 155.9, 220.9, 335.5]
apl_stock_prices = [20.4, 17.5, 25.7, 25.4, 27.4, 42.1, 52.5, 69.9, 127.5, 151.2]
years = np.arange(2012, 2022)
```

```python
plt.figure(figsize=(8,4))
plt.title('Explanatory title')
plt.plot(years, ms_stock_prices, label='MS stock')
plt.plot(years, apl_stock_prices, label='Apple stock')
plt.legend()
plt.xlabel('years')
plt.ylabel('stock price')
plt.xticks(years[::3], rotation=45)
plt.show()
```

###Subplots

```python
fig_1 = plt.figure(figsize=(16,4))
ax_1 = fig_1.add_subplot(1, 3, 1)
ax_2 = fig_1.add_subplot(1, 3, 2)
ax_3 = fig_1.add_subplot(1, 3, 3)
ax_2.plot(years, ms_stock_prices)
```

```python
fig_1 = plt.figure(figsize=(6,6))
ax_1 = fig_1.add_subplot(2, 2, 1)
ax_2 = fig_1.add_subplot(2, 2, 2)
ax_3 = fig_1.add_subplot(2, 2, 3)
ax_4 = fig_1.add_subplot(2, 2, 4)
ax_1.plot(years, ms_stock_prices, color='red', label='MS stock')
ax_4.scatter(years, apl_stock_prices, color='green', label='Apple stock')
ax_1.legend()
plt.show()
```

**Ex: Write a function that returns the maximum value in a list**

```python
def find_maximum(list_of_numbers):
    maximum = list_of_numbers[0]
    for number in list_of_numbers:
        if number > maximum:
            maximum = number

    return maximum
```

```python
find_maximum([0,-2, 30, -45, 90])
```

```python
def find_maximum(list_of_numbers):
    sorted_list = sorted(list_of_numbers)
    return sorted_list[-1]
```

```python
find_maximum([0,-2, 30, -45, 90])
```
