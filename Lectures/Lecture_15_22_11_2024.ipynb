##Exercise: find player that has the best PLUS_MINUS on average (mean):

```python
games_details_df_groupby_PLAYER_PLUS_MINUS = games_details_df.groupby('PLAYER_NAME')['PLUS_MINUS']
```

```python
max_plus_minus_on_average = games_details_df_groupby_PLAYER_PLUS_MINUS.mean().sort_values(ascending=False)[0]
best_minus_mask = games_details_df_groupby_PLAYER_PLUS_MINUS.mean() == max_plus_minus_on_average
```

```python
best_minus_mask.sum()
```

```python
games_details_df_groupby_PLAYER_PLUS_MINUS.mean().sort_values(ascending=False)[best_minus_mask]
```
